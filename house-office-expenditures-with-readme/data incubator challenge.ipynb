## Question 3

What was the highest average staff salary among all representatives in 2016? Assume staff sizes is equal to the number of unique payees in the 'PERSONNEL COMPENSATION' category for each representative.

```python
total_payments=pd.DataFrame()
cnt=0
for x in range(1,len(names)):
    print(x)
    cnt=x
    i=names[x]
    
    print(i)
    if cnt == 31:
        payments=pd.read_csv(i,encoding = "ISO-8859-1")
        
        payments['AMOUNT']=payments['PURPOSE']
        
        mask_positive=payments['AMOUNT']>0
        payments=payments.loc[mask_positive]

        mask_empty=payments['START DATE'].str.isspace()
        payments=payments.loc[~mask_empty]


        payments['START DATE']=pd.to_datetime(payments['START DATE'])
        payments['END DATE']=payments['PAYEE']
        
        payments['END DATE']=pd.to_datetime(payments['END DATE'])
        payments['COVERAGE PERIOD']=payments['END DATE']-payments['START DATE']
        print(payments.shape)
            
        total_payments= pd.concat([total_payments,payments], ignore_index=True)
        
        print(total_payments.shape)

        print(cnt)
        cnt=cnt+1
        
    elif 32 <= x < 34:
        
        payments=pd.read_csv(i, thousands=',')
        mask_positive=payments['AMOUNT']>0
        payments=payments.loc[mask_positive]
        
        mask = payments['START DATE'].str.contains('   ', case=False, na=False)

        payments=payments.loc[~mask]
        
        payments['START DATE']=pd.to_datetime(payments['START DATE'])
        payments['END DATE']=pd.to_datetime(payments['END DATE'])
        
        payments['COVERAGE PERIOD']=payments['END DATE']-payments['START DATE']
        
        total_payments= pd.concat([total_payments,payments], ignore_index=True)
        
        cnt=cnt+1
        
        
    elif cnt == 34:
        
        payments=pd.read_csv(i,encoding = "ISO-8859-1")
        mask_positive=payments['AMOUNT']>0
        payments=payments.loc[mask_positive]

        mask_empty=payments['START DATE'].str.isspace()
        payments=payments.loc[~mask_empty]


        payments['START DATE']=pd.to_datetime(payments['START DATE'])
        payments['END DATE']=pd.to_datetime(payments['END DATE'])
        payments['COVERAGE PERIOD']=payments['END DATE']-payments['START DATE']
        
        print(payments.shape)
        
        #mask = (payments['START DATE'] > '2010-01-01') & (payments['START DATE'] <= '2016-12-31')
        #payments=payments.loc[mask]
        total_payments= pd.concat([total_payments,payments], ignore_index=True)
        print(total_payments.shape)
        
        
       
        print(cnt)
        cnt=cnt+1
        
    else:
        payments=pd.read_csv(i, thousands=',')
        mask_positive=payments['AMOUNT']>0
        payments=payments.loc[mask_positive]
        payments['START DATE']=pd.to_datetime(payments['START DATE'])
        payments['END DATE']=pd.to_datetime(payments['END DATE'])
        payments['COVERAGE PERIOD']=payments['END DATE']-payments['START DATE']
        
        print(payments.shape)

        total_payments= pd.concat([total_payments,payments], ignore_index=True)
        print(total_payments.shape)
        
        print(cnt)
        cnt=cnt+1
```
